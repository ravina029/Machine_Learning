# DBSCAN algorithm 


```python
from sklearn.cluster import DBSCAN
import numpy as np
import matplotlib.pyplot as plt


```

```python
# Create a sample dataset
X = np.array([[1, 2], [2, 2], [2, 3], [8, 7], [8, 8], [25, 80]])

# Apply DBSCAN
dbscan = DBSCAN(eps=3, min_samples=2)
labels = dbscan.fit_predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', edgecolors='k')
plt.title('DBSCAN Clustering')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# Access cluster labels
print("Cluster Labels:", labels)
```

# observation:
here data is divided into three clusters depending on the parameters we supplied. 

# Let's observe another dataset

```python
# Create a dataset forming a smiley pattern
np.random.seed(0)
theta = np.linspace(0, 2*np.pi, 500)  # Increase the number of points for a denser pattern
smiley_pattern = np.array([np.cos(theta), np.sin(theta)]).T + 0.1 * np.random.randn(500, 2)
```

```python
# Add noise points
noise = 0.5 * np.random.randn(100, 2)  # Increase the number of noise points
dataset = np.concatenate([smiley_pattern, noise])
```

```python
# Apply DBSCAN
dbscan = DBSCAN(eps=0.15, min_samples=3)
labels = dbscan.fit_predict(dataset)
```

```python
# Plot the results
plt.figure(figsize=(8, 8))
plt.scatter(dataset[:, 0], dataset[:, 1], c=labels, cmap='viridis', edgecolors='k')
plt.title('DBSCAN Clustering - Smiley Pattern')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# we can observe different outcomes by changing the parameters of the algorithm
```
