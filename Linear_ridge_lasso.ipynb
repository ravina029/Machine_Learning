```python
#linear regression 
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
lin_reg= LinearRegression()
```

```python
CV=[5,6,7,8]
Mse=[]
for i in CV:
    mse=cross_val_score(lin_reg,X,y,scoring='neg_mean_squared_error',cv=i)
    mean_mse=np.mean(mse)
    Mse.append((mean_mse))
    print(f"CV={i} and mse={mean_mse}")
print(Mse)
print(f"the best score is {np.sort(Mse)[-1]}")
```

## Logistic regression

```python
## Logistic regression
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_breast_cancer
```

```python
df=load_breast_cancer()
X=pd.DataFrame(df['data'],columns=df['feature_names'])
X.head()
```

```python
X.describe()
```

```python
y=pd.DataFrame(df['target'],columns=["Target"])
y
```

```python
y['Target'].value_counts()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=34)
```

```python
params=[{'C':[1,5,10]},{'max_iter': [100,150,200]}]
```

```python
model=LogisticRegression(C=10,max_iter=150)
```

```python
model1=GridSearchCV(model,param_grid=params, scoring='f1',cv=5)
```

```python
model1.fit(X_train,y_train)
```

```python
model1.best_params_
```

```python
model1.best_score_
```

```python
y_pred=model1.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report,accuracy_score
```

```python
confusion_matrix(y_pred,y_test)
```

```python
accuracy_score(y_pred,y_test)
```

```python
print(classification_report(y_pred,y_test))
```
