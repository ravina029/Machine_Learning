#We can see that the data to be classified is imbalanced, Let's oversample the data using SMOTE.
SMOTE stands for Synthetic Minority Over-sampling Technique. It is a popular technique in machine learning and data preprocessing, particularly in the context of imbalanced datasets. Imbalanced datasets occur when one class in a classification problem has significantly fewer samples than the other class(es), making it challenging for machine learning models to learn and predict the minority class effectively.

```python
from imblearn.over_sampling import SMOTE
sm=SMOTE(random_state=1)
X,Y=sm.fit_resample(X,Y)
```

```python
#Scaling the data
from sklearn.preprocessing import MinMaxScaler
MS=MinMaxScaler()
X=MS.fit_transform(X)
```

```python
X
```

```python
#Train_test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,Y, test_size=0.2, random_state=34)
```

# Model development

```python
#Algorithm 1
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
from math import sqrt 
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
parameters={"n_estimators":[50,70,100,120,150,200],"max_features":["sqrt","auto"],
           "max_depth":[2,4,6],"min_samples_split":[2,3,4,5],
           "min_samples_leaf":[1,2,3]}
clf=GridSearchCV(rf,parameters,cv=6,scoring="accuracy")
clf.fit(X_train,y_train)
clf.best_params_
```

```python
#After parameter tuning
rf=RandomForestClassifier(max_depth= 6,
 max_features= 'sqrt',
 min_samples_leaf= 1,
 min_samples_split= 3,
 n_estimators= 150)
rf.fit(X_train,y_train)
y_pred=rf.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python

print(confusion_matrix(y_test,y_pred))
```

```python

print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier 
clf=DecisionTreeClassifier() 
clf.fit(X_train,y_train) 
y_pred1=clf.predict(X_test) 
print(classification_report(y_test,y_pred1))
```

```python
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred1))
```

```python
#algorithm 3, Adaboost without hyperparameter_tuning
from sklearn.ensemble import AdaBoostClassifier 
ab=AdaBoostClassifier() 
ab.fit(X_train,y_train) 
y_pred3=ab.predict(X_test) 
print(classification_report(y_test,y_pred3))
```

```python
#parameter tuning
params2={'n_estimators':[30,50,70,90,100,120,150,170,200,240,250],'algorithm':['SAMME','SAMME.R'],
         'random_state':[1,2,3,4,5],'learning_rate':[0.5,1.0,2.0,3.0,4.0]} 
clf=GridSearchCV(ab,params2,cv=6,scoring='accuracy') 
clf.fit(X_train,y_train)
clf.best_params_
```

```python
#After hyperparameter tuning
ab=AdaBoostClassifier(algorithm='SAMME',learning_rate=1.0 ,n_estimators=150,random_state=1)
ab.fit(X_train,y_train)
y_pred5=ab.predict(X_test)
print(classification_report(y_test,y_pred5))
```

```python
#ALGORITHM 4: GRADIENT BOOSTING CLASSIFIER
from sklearn.ensemble import GradientBoostingClassifier 
gb=GradientBoostingClassifier(n_estimators=500,learning_rate=0.4,max_depth=8,random_state=1) 
gb.fit(X_train,y_train) 
y_pred4=gb.predict(X_test) 
print(classification_report(y_test,y_pred4))
```

```python
#ALGORITHM 5: NAIVE BAYES
from sklearn.naive_bayes import GaussianNB 
nb=GaussianNB()
nb.fit(X_train,y_train)
y_pred5=nb.predict(X_test) 
print(classification_report(y_test,y_pred5))
```

```python
#ALGORITHM 6 : K-NEIGHBORS
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report,ConfusionMatrixDisplay 
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,y_train)
y_pred6=knn.predict(X_test)
print(classification_report(y_test,y_pred6))
```

```python
#ALGORITHM 7: SVC(without hyperparameter tuning)
from sklearn.svm import SVC
sv=SVC()
sv.fit(X_train,y_train)
y_pred7=sv.predict(X_test) 
print(classification_report(y_test,y_pred7))
```

```python
#After hyperparameter tuning
from sklearn.model_selection import GridSearchCV
sv=SVC() 
params={'C':[0.1,1,2,3,4,10,100],'gamma':[1,0.1,0.5,0.01,0.001],'kernel':['rbf','poly','sigmoid']} 
clf=GridSearchCV(sv,params,cv=5,scoring='accuracy') 
clf.fit(X_train,y_train)
clf.best_params_
```

```python

sv=SVC(C=100,kernel='poly',gamma=1)
sv.fit(X_train,y_train)
y_pred=sv.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
#ALGORITHM 8: NAIVE BAYES
from sklearn.naive_bayes import GaussianNB 
nb=GaussianNB()
nb.fit(X_train,y_train)
y_pred8=nb.predict(X_test) 
print(classification_report(y_test,y_pred8))
```

# Save the model

we will save our model usin pickle module: A pickle file is a serialized file format used in Python for storing and exchanging complex data structures, including objects, dictionaries, lists, and more. Pickling is the process of converting a Python object into a byte stream, and unpickling is the process of recreating the original Python object from the byte stream. Pickle files have the .pkl extension and are often used for tasks like saving the state of a program or sharing data between Python scripts.

```python
import pickle 
with open('sv.pkl', 'wb') as file:
    pickle.dump(sv, file)

print("Model saved to model.pkl")
```

```python
with open('sv.pkl', 'rb') as file:
    loaded_model = pickle.load(file)
```

```python
input_data=(1,9,89.1,6,500,6,3,77,420,126,83)
input_numpy=np.asarray(input_data) 
input_data_reshape=input_numpy.reshape(1,-1)
prediction=loaded_model.predict(input_data_reshape)
print(prediction)
if(prediction[0]==0):
    print('this person has insomnia')
elif(prediction[0]==1):
    print('this person is alright') 
else:
    print('this person has sleep apnea')
```
