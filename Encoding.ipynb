```python
df.head()
```

# One hot encoding using pandas function get_dummies

```python
#It is clear from the data that "Brand" and "fuel" are categorical,
# but nominal that doesn' contain any specific order. owner is odered data. 

pd.get_dummies(df,columns=["fuel","owner"])
```

```python
X_train_new = ohe.fit_transform(X_train[['fuel','owner']])
```
