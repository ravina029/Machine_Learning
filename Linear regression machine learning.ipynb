# linear regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score #it will do the random train test split to learn in a best way
model1=LinearRegression()
mse=cross_val_score(model1,X_train,y_train,scoring="neg_mean_squared_error",cv=5)
```

# Ridge regression 

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
ridge=Ridge()
params={"alpha":[1e-15,1e-10,1e-8,1e-7,1e-3,1e-2,1,2,3,5,10,20,39,26,40,50,60,70,80,90,100]} #before applying the gridsearchcv lets apply these alpha
ridge_regressor=GridSearchCV(ridge,params,scoring="neg_mean_squared_error",cv=5)
ridge_regressor.fit(X_train,y_train)
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
```

```python
lasso=Lasso()
params={"alpha":[1e-15,1e-10,1e-8,1e-7,1e-3,1e-2,1,2,3,5,10,20,39,26,40,50,60,70,80,90,100]} #before applying the gridsearchcv lets apply these alpha
lasso_regressor=GridSearchCV(lasso,params,scoring="neg_mean_squared_error",cv=5)
lasso_regressor.fit(X_train,y_train)
```

```python
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```

```python
y_pred=lasso_regressor.predict(X_test)
from sklearn.metrics import r2_score
```

```python
from sklearn.metrics import r2_score
r2_score1=r2_score(y_pred,y_test)
```
